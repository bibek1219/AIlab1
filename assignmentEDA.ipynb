```python
import pandas as pd
import seaborn  as sns
import matplotlib.pyplot as plt
```

```python
# Creating a pandas dataframe 
column_name = ['sepal length','sepal width','petal length','petal width','class']
iris_df = pd.read_csv('C:/Users/ACER/Machine Learning/Dataset/Iris dataset/Iris.csv',header = None, names = column_name)
iris_df
```

```python
iris_df.shape
```

```python
# Showing the category of the data
sns.countplot(data=iris_df, x='class')
iris_df['class'].value_counts()
```

```python
# Pie chart
iris_df['class'].value_counts().plot(kind='pie',autopct='%0.2f')
```

```python
# Histogram of the data
iris_df.hist(figsize = (10,10))
```

```python
# Distplot
sns.distplot(iris_df['sepal length'],hist=False)
```

```python
# Boxplot (5 point summary)
sns.boxplot(iris_df)
```

```python
# Pairplot
sns.pairplot(iris_df,hue='class')
```

```python
from sklearn import preprocessing
label_encoder = preprocessing.LabelEncoder()
#Converting categorical data into numerical value
iris_df['label']= label_encoder.fit_transform(iris_df['class'])
iris_df
```

```python
# Dropping class column
iris_df1 = iris_df.drop(['class'], axis = 1)
iris_df1
```

```python
# Heatmap
plt.figure(figsize =(5,5))
sns.heatmap(iris_df1.corr(), annot = True)
```
